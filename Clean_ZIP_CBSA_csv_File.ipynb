```python
# Dependencies
import pandas as pd
import os
import csv
```

```python
# Store filepath into a variable
filepath = "Data/ZIP_CBSA_122021_csv.csv"
```

```python
#Read out file with Pandas
df = pd.read_csv(filepath, encoding="UTF-8")
```

```python
# Show header
df.head()
```

```python
df.head()
```

```python
df.reset_index(drop=True)
```

```python
# Remove extraneous columns
del df['usps_zip_pref_city']
del df['usps_zip_pref_state']
del df['res_ratio']
del df['bus_ratio']
del df['oth_ratio']
del df['tot_ratio']
df.head()
```

```python
# View data type of two remaining columns
#df['zip'].dtypes
#df['cbsa'].dtypes

#Update Zip and CBSA Column to use VARCHAR Datatype
df = df.astype({"zip": str})
df = df.astype({"cbsa": str})

# View updated data type of two remaining columns
#df['zip'].dtypes
#df['cbsa'].dtypes
```

```python
# leading edge zeroes fix - in line
df['zip'] = df['zip'].apply(lambda x: x.zfill(5))
df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df.to_csv("cleaned_CSVs/CleanedZIP_CBSA_2022_04_05.csv", index=False, header=True)
```
