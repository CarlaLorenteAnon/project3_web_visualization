```python
# Import Dependencies
import pandas as pd
import csv
import os
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States'
browser.visit(url)
```

```python
tables = pd.read_html(url)
tables
```

```python
type (tables)
```

```python
df = tables[1]
df.head()
```

```python
to_drop = ['Recreation visitors (2021)[11]', 'Area (2021)[13]', 'Description', 'Image']
df.drop(to_drop, inplace=True, axis=1)
df.head()
```

```python
new = df["Location"].str.split("/", n = 2, expand = True)
  
df["Coordinates"]= new[1]
  
df.head()
```

```python
to_drop = ['Location', 'Initial Coordinates']
df.drop(to_drop, inplace=True, axis=1)
df.head()
```

```python
df.to_csv ("nationalParksInitial.csv", index = False, header=True)
```
